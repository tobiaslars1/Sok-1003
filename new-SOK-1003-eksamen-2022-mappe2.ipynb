a) Du har nå begynt å spille lotto i stedet, og satser alt på ett vinnertall. Lag en while-løkke som printer ut tall helt til du har trukket riktig tall (som du definerer selv). For enkelthets skyld kan du begrense utfallsrommet av trekningene til mellom 0-30.

```python
lottery = np.random.randint(0, 30) # Definerer verdiene det skal trekkes mellom.

lottery_number = 23 # Definerer lotto nummer.

tries = 0 # Antall forsøk.

lottery_numbers = []

while lottery != lottery_number :
    if lottery != lottery_number : # Hvis det ikke er lotto nummer så skal den plusse på 1 på tries.
        tries += 1
        lottery = np.random.randint(0, 30)
        
    lottery_numbers.append(lottery)
        
print(lottery_numbers) # Printer alle tallene.
    
print("it took", tries, "tries to get", lottery_number) # Printer en setning, som også har tries i seg.
```

### Oppgave 3 (20 poeng)

En bedrift produserer biler. Produktfunksjonen til bedriften defineres slik  $f(L, a, R) = 2RL^{a}$, hvor:
- `L` er arbeidskraft, 
- `a` er produktiviteten til arbeiderne og 
- `R` er antall robotmaskiner

a) Lag en formel for produktfunksjonen til bedriften og plot den grafisk med ulike verdier av `L` på x-aksen. Anta `a=0.6` og `R=2`

```python
def productfunc(L, a, R): # Definerer produktfunksjonen
    return 2*R*L**a

a = 0.6 # Legger inn verdiene
R = 2
L = np.linspace(0, 15000) # Angir x akse

fig3, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.plot(L, productfunc(L, a, R))
plt.title("Productivity for workers and robots in car manufacturing") # Navngir akser og tittel
plt.xlabel("Robots and workers")
plt.ylabel("Cars manufactured")
plt.style.use("seaborn-whitegrid") # Style
    
plt.show(fig3) # Viser figur
```

b) anta at profittfunksjonen til denne bedriften er ${profit = f(L, a, R)p-wL-cR-K}$, hvor 
- `w` er månedslønnen til arbeiderne, 
- `c` er kostnaden for robotmaskinene 
- `K` er faste kostnader
- `p` er utsalgsprisen på bilene. 

Anta `a=0.6`,`R=6`, `p=300 000`, `w=100 000`, `c= 1 000 000` og `K=90 000 000`. Plot profittfunksjonen figurativt for antall arbeidere (`L`) mellom 0 og 10 000. Vis profitten i millioner (dvs at du må dele på 1 000 000)

```python
def profit(L, a, R, p, w, c, K): # Definerer profitt, med produktfunc og de nye variablene
    return productfunc(L, a, R) * p - w * L - c * R - K

# Legger til de nye tallene
L = np.linspace(0, 15000)
R = 6 
p = 300000
w = 100000
c = 1000000
K = 90000000


fig4, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.plot(L, profit(L, a, R, p, w, c, K)/1000000)
plt.title("Profit for car manufacturers")
plt.xlabel("Robots and workers")
plt.ylabel("Profit (in millions)")
plt.style.use("seaborn-whitegrid")

plt.show(fig4)
```

c) Plot profitttfunksjonen for antall robostmaskiner `R=[3, 6, 9]` i samme plot (dvs at tre profittfunksjoner vises sammen). Bruk av "for loops" for å gjøre dette belønnes

```python
R = [3, 6, 9] # Endrer R fra 3 til [3,6,9]

fig5, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.xlabel("Robots and workers")
plt.ylabel("Profit (in millions)")
plt.title("Profit for car manufacturers")

for i in R: # For loop, og i profit bytter jeg R til i.
    plt.plot(L, profit(L, a, i, p, w, c, K)/1000000, label = f"{i} robots")
    plt.legend()
```

d) finn profittmaksimum og optimal antall arbeidere ved hjelp av derivasjon med samme forutsetninger som i (1b). Bruk `sympy`-pakken til dette 

```python
# Oppskrift fra 5. sympy
L,a,R,p,w,c,K=sp.symbols("L a R p w c K") # Definerer symbol

d_profitt=sp.diff(profit(L,a,R,p,w,c,K),L) # sp.diff av symbolene

foc=sp.Eq(d_profitt,0) # Definerer foc eller "førsteordensbetingelsen"

L_max=solve(foc,L)[0]

profit_max=profit(L_max,a,R,p,w,c,K) # Finner ut hvor profitt er på maks.

num_dict={a:0.6,w:100000,p:300000,K:90000000,c:1000000,R:6} # Legger til verdiene vi vet

L_max.subs(num_dict)

num_dict[L]=L_max.subs(num_dict)

profit_max_num=float(profit(L,a,R,p,w,c,K).subs(num_dict))
profit_max_num


tbl=f"""
|                                 | Decimal value                             |  Analytical value              | 
| :-------------------------------| :----------------------------------------| :-----------------------------| 
| Optimal workforce:     | ${np.round(float(num_dict[L]),1)}$         |${sp.latex(L_max)}$            | 
| Max profit                | ${np.round(float(profit_max_num),1)}$    |${sp.latex(profit_max)}$       | 

"""
display(Markdown(tbl))
```

e) vis figurativt med bruk av `fill_between` arealet hvor man taper penger (i rødt) og hvor man tjener penger (i grønt). Marker også profittmaksimum og antall arbeidere i profittmaksimum - gjerne ved bruk av `vlines`. Bruk ellers samme forutsetninger for argumentene som i oppgave (1b)

```python
def profit(L, a, R, p, w, c, K): # Definerer profitt, med produktfunc og de nye variablene
    return productfunc(L, a, R) * p - w * L - c * R - K

# Legger til de nye tallene
L = np.linspace(0, 15000)
R = 6 
p = 300000
w = 100000
c = 1000000
K = 90000000
line = 0



fig6, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.plot(L, profit(L, a, R, p, w, c, K)/1000000)
plt.scatter(2168.4, 48.5, color = "red", label = "Max profit 48.5M")
plt.vlines(2168.4, -400, 48.5, linestyle="dashed", color = "black", label = "Optimal workforce 2168")
plt.axhline(y = 0, linestyle = "solid", color = "green", alpha = 1)
plt.fill_between(L, profit(L, a, R, p, w, c, K)/1000000, where = profit(L, a, R, p, w, c, K)/1000000 > line,  # kode hentet fra tutorialspoint.
                 color = "green", alpha = 1)
plt.fill_between(L, profit(L, a, R, p, w, c, K)/1000000, where = profit(L, a, R, p, w, c, K)/1000000 < line,  
                 color = "red", alpha = 1)
plt.legend()

plt.show(fig6)
```
