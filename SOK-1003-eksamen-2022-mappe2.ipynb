### Oppgave 2 (10 poeng)
a) Du har nå begynt å spille lotto i stedet, og satser alt på ett vinnertall. Lag en while-løkke som printer ut tall helt til du har trukket riktig tall (som du definerer selv). For enkelthets skyld kan du begrense utfallsrommet av trekningene til mellom 0-30.

```python
lottery = np.random.randint(0, 30) # Definerer verdiene det skal trekkes mellom.

lottery_number = 23 # Definerer lotto nummer.

tries = 0 # Antall forsøk.

while lottery != lottery_number :
    if lottery != lottery_number : # Hvis det ikke er lotto nummer så skal den plusse på 1 på tries.
        tries += 1
        lottery = np.random.randint(0, 31)
        
    print(lottery) # Printer alle tallene.
    
print("it took", tries, "tries to get", lottery_number) # Printer en setning, som også har tries i seg.
        
```

### Oppgave 3 (20 poeng)

En bedrift produserer biler. Produktfunksjonen til bedriften defineres slik  $f(L, a, R) = 2RL^{a}$, hvor:
- `L` er arbeidskraft, 
- `a` er produktiviteten til arbeiderne og 
- `R` er antall robotmaskiner

a) Lag en formel for produktfunksjonen til bedriften og plot den grafisk med ulike verdier av `L` på x-aksen. Anta `a=0.6` og `R=2`

```python
import matplotlib.pyplot as plt

def productfunc(L, a, R): # Definerer produktfunksjonen.
    return 2*R*L**a

a = 0.6
R = 2
L = np.linspace(0, 15000)

fig3, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.plot(L, productfunc(L, a, R))
plt.title("Productivity for workers and robots in car manufacturing")
plt.xlabel("Robots and workers")
plt.ylabel("Cars manufactured")
plt.style.use("seaborn-whitegrid")
    
plt.show(fig3)
```

b) anta at profittfunksjonen til denne bedriften er ${profit = f(L, a, R)p-wL-cR-K}$, hvor 
- `w` er månedslønnen til arbeiderne, 
- `c` er kostnaden for robotmaskinene 
- `K` er faste kostnader
- `p` er utsalgsprisen på bilene. 

Anta `a=0.6`,`R=6`, `p=300 000`, `w=100 000`, `c= 1 000 000` og `K=90 000 000`. Plot profittfunksjonen figurativt for antall arbeidere (`L`) mellom 0 og 10 000. Vis profitten i millioner (dvs at du må dele på 1 000 000)

```python
def profit(L, a, R, p, w, c, K): # Definerer profitt, de nye variablene, samt variablene fra productfunc().
    return productfunc(L, a, R) * p - w * L - c * R - K

L = np.linspace(0, 15000)
R = 6
p = 300000
w = 100000
c = 1000000
K = 90000000


fig4, ax = plt.subplots(figsize = (7.5, 4), dpi = 100) # Plotter dette inn i en graf.
plt.plot(L, profit(L, a, R, p, w, c, K)/1000000)
plt.title("Profit for car manufacturers")
plt.xlabel("Robots and workers")
plt.ylabel("Profit (in millions)")
plt.style.use("seaborn-whitegrid")

plt.show(fig4)
```

c) Plot profitttfunksjonen for antall robostmaskiner `R=[3, 6, 9]` i samme plot (dvs at tre profittfunksjoner vises sammen). Bruk av "for loops" for å gjøre dette belønnes

```python
R = [3, 6, 9] # Lager R med forskjellige values.

fig5, ax = plt.subplots(figsize = (7.5, 4), dpi = 100)
plt.xlabel("Robots and workers")
plt.ylabel("Profit (in millions)")
plt.title("Profit for car manufacturers")

for i in R:
    plt.plot(L, profit(L, a, i, p, w, c, K)/1000000) # Bytter R ut med i, slik at det lages tre linjer.
```
